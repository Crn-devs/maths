# The Law of exponents
**When multiplying powers with the same base, add the exponents.**

- 3^3 x 3^8 = 3^11

- 3^3 x 4^5 = 3x3x3 x 4x4x4x4 -> here the exponents cannot be added due to the base being different



```go
import 	("github.com/janpfeifer/gonb/gonbui"
		"math"
		"fmt"
		"github.com/gonb/plot"
    	"github.com/gonb/ui"
)
```

```go
%%

// there are a few ways to compute Exponents in Go

// manually
result1 := 3*3*3
eq := fmt.Sprintf(`$%v^%v = %v$`, 3.0, 3.0, result1)
gonbui.DisplayMarkdown(eq)

// Within a loop manually (alot slower than built ins like math.Pow)
exponent := 5.00
base := 3.00
result := 1.0
for i := 0.0; i < exponent; i++ {
	result *= base
}
eq2 := fmt.Sprintf(`$%v^%v = %v$`, base, exponent, result)
gonbui.DisplayMarkdown(eq2)

eq3 := fmt.Sprintf(`$%v^%v = %v$`, 3.0, 3.0, math.Pow(3,3))
gonbui.DisplayMarkdown(eq3)

```

```go
%%

base1 := 3.00
exponent := 5.00

result := math.Pow(base1,exponent) // built in for exponents in go vs using a loop
eq := fmt.Sprintf(`$ %v^%v = %v $`, base1,exponent,result)

gonbui.DisplayMarkdown(eq)
```

```go
%%

base1 := 3.00
base2 := 4.00
exponent1 := 5.00
exponent2 := 3.00

if base1 != base2 {
	result1 := math.Pow(base1,exponent1) 
	result2 := math.Pow(base2,exponent2) 

	eq := fmt.Sprintf(`$ %v^%v \times %v^%v = %v $`, base1,exponent1,base2, exponent2, result1+result2)
	gonbui.DisplayMarkdown(eq)
	return
}

exponent := exponent1+exponent2 // adding expoenents
result := math.Pow(base1,exponent)
eq := fmt.Sprintf(`$ %v^%v\times%v^%v = %v $`, base1,exponent1,base2, exponent2, result)

gonbui.DisplayMarkdown(eq)
```

```go
%%
// computing squareroot using fractional powers 
fmt.Println(math.Pow(4, 0.5))
```

```go
%%
x := 5.0
y := 5.1

result := math.Pow(x , 0.75) * math.Pow(4.0, y)
resultRounded := fmt.Sprintf("%.2f\n", result)

eq := fmt.Sprintf(`$ %v^\frac{3}{4}\times%v^{%v} = %v $`, x,4.0, y, resultRounded)

gonbui.DisplayMarkdown(eq)
```

```go
%%
x := 5.0
y := 5.1

result := math.Pow(3.0 , 3.0) / math.Pow(x, y)
resultRounded := fmt.Sprintf("%.08f\n", result)

eq := fmt.Sprintf(`$ \frac{3^3}{x^y} = %v $`,resultRounded)

gonbui.DisplayMarkdown(eq)
```

```go
%%
x := 5.0
// y := 5.1

result := math.Pow(10.0 , x-4)
eq := fmt.Sprintf(`$ 10^{x-4} = %v $`,result)

gonbui.DisplayMarkdown(eq)
```

Add parenthesis to make the following expression true

2. 9 / 3 + 6 - 1 = 0

```go

%%
eq1 := fmt.Sprintf(`$ 4 \times 5 \div 7 + 3 = %v $`, (4*5) / (7+3)) // 4x5 / 7+3
eq2 := fmt.Sprintf(`$ 9 \div 3 + 6 - 1 = %v $`, 9/(3+6)- 1) // 9/9 = 1 -1

gonbui.DisplayMarkdown(eq1)
gonbui.DisplayMarkdown(eq2)
```
