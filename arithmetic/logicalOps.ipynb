# Logical operations in python

Exercise 
    loop over number 0-3 (i)
        nested loop over numbers 0-4 (j)

    print i-j for numbers where i>0, j>0

```python
from IPython.display import display,Math

for i in range(4):
    for j in range(5):
        # if op + and op + > greater than
        if i>0 and j>0:
            display(Math('%g^{-%g} = %g' % (i, j, i**-j)))
```
