Checking if a number is even or odd using modulus

if a number can be divided by 2 and have no remainder it is positive

```python
numbers = range(-5, 6)

for v in numbers:
    if v%2 == 0:
        print(str( v) + " is even")
    else:
        print(str(v) + " is odd")
```
